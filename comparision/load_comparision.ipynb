# Add To Cart Response Time

## Old Architecture

```python
import json

# file from k6 run --out json=results.json
filename = 'old_architecture\\add_to_cart.json'

durations = []

with open(filename, 'r') as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        # each line is a JSON object
        obj = json.loads(line)
        # check if this is the metric we want
        if obj.get('metric') == 'place_order_duration_ms':
            if 'value' not in obj['data']:
                continue
            value = obj['data']['value']
            durations.append(value)

print(f"Collected {len(durations)} samples")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Calculate statistics
median = np.median(durations)
mean = np.mean(durations)
p95 = np.percentile(durations, 95)
p99 = np.percentile(durations, 99)

# X-axis: request index
requests = np.arange(1, len(durations) + 1)

# Plot vertical bars
plt.figure(figsize=(15,6))
plt.bar(requests, durations, color='lightgray', edgecolor='gray')

# Plot median, mean, p95, p99 lines
plt.axhline(y=median, color='blue', linestyle='--', linewidth=1.5, label='Median (p50)')
plt.axhline(y=mean, color='green', linestyle='-', linewidth=1.5, label='Mean')
plt.axhline(y=p95, color='orange', linestyle='--', linewidth=1.5, label='95th percentile')
plt.axhline(y=p99, color='red', linestyle='--', linewidth=1.5, label='99th percentile')

plt.xlabel('Request Index')
plt.ylabel('Response Time (ms)')
plt.title('Response Time Distribution per Request')
plt.legend()
plt.show()
```
